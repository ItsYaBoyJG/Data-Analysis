```python
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
```

```python
URL = "https://www.nfl.com/stats/team-stats/offense/passing/2024/reg/all"

page = requests.get(URL)
html = page.content
soup = BeautifulSoup(html, "html.parser")
```

```python
table_name = "d3-o-table d3-o-table--detailed d3-o-team-stats--detailed d3-o-table--sortable {sortlist: [[0,0]], sortinitialorder: 'asc'}"

table = soup.find("table", class_=table_name)

print(table)
```

```python
columns = soup.findAll('th')

print(columns)
```

```python
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '',text)

titles = ()

for i in columns:
    x = remove_html_tags(str(i))
    l = list(titles)
    if x != "None":
        t = x.replace('\n', '')
        l.append(t)
        titles = tuple(l)

print(titles)
```
