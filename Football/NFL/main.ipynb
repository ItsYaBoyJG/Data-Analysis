```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as mpl
import re

```

```python
YEAR = '2023'
#rushing receiving scoring downs tackles 
# fumbles interceptions field-goals sociring 
# kickoff kickoff-returns punting punt-returns
STAT_TYPE = 'passing' 
OFFENSE_DEFENSE = 'offense' #defense

URL = "https://www.nfl.com/stats/team-stats/offense/passing/2024/reg/all"

page = requests.get(URL)
html = page.content
soup = BeautifulSoup(html, "html.parser")
```

```python
table_name = "d3-o-table d3-o-table--detailed d3-o-team-stats--detailed d3-o-table--sortable {sortlist: [[0,0]], sortinitialorder: 'asc'}"

table = soup.find("table", class_=table_name)

print(table)
```

```python
columns = soup.findAll('th')

print(columns)
```

```python
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '',text)

titles = ()

for i in columns:
    x = remove_html_tags(str(i))
    l = list(titles)
    if x != "None":
        t = x.replace('\n', '')
        l.append(t)
        titles = tuple(l)

print(titles)
```

```python
df = pd.DataFrame(columns=['Team', 'Att', 'Cmp','Cmp %', 'Yds/Att', 'Pass Yds','TD', 'INT', 'Rate', '1st', '1st%', '20+', '40+', 'Lng', 'Sck', 'SckY'])
```

```python
n = 'd3-o-club-shortname'

for row in table.find_all('tr'):
    c = row.find_all('td')
    
    
    if (c != []):
        n = remove_html_tags(str(c[0])).replace('\n','')
        name = remove_html_tags(str(c[0])).replace('\n','')
        att = remove_html_tags(str(c[1])).replace('\n','')
        comp = remove_html_tags(str(c[2])).replace('\n','')
        cmpPer = remove_html_tags(str(c[3])).replace('\n','')
        ydsPatt = remove_html_tags(str(c[4])).replace('\n','')
        pYds = remove_html_tags(str(c[5])).replace('\n','')
        td = remove_html_tags(str(c[6])).replace('\n','')
        interceptions = remove_html_tags(str(c[7])).replace('\n','')
        rate = remove_html_tags(str(c[8])).replace('\n','')
        first = remove_html_tags(str(c[9])).replace('\n','')
        firstPerc = remove_html_tags(str(c[10])).replace('\n','')
        twentyP = remove_html_tags(str(c[11])).replace('\n','')
        fourtyP = remove_html_tags(str(c[12])).replace('\n','')
        lng = remove_html_tags(str(c[13])).replace('\n','')
        sck = remove_html_tags(str(c[14])).replace('\n','')
        sckY = remove_html_tags(str(c[15])).replace('\n','')
        
        
        df = pd.concat([df, pd.DataFrame.from_records([{'Team': name, 'Att': att, 'Cmp': comp, 'Cmp%': cmpPer, 'Yds/Att': ydsPatt, 'Pass Yds': pYds, 'TD': td, 'INT': interceptions, 'Rate':rate, '1st': first, '1st%': firstPerc,'20+': twentyP, '40+': fourtyP, 'Lng': lng, 'Sck': sck, 'SckY': sckY}])], ignore_index=True)


df.head()    
```
