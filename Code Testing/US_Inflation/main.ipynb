```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as mpl
import matplotlib.dates as mdates
import re
```

```python
URL = 'https://www.usinflationcalculator.com/inflation/current-inflation-rates/'

page = requests.get(URL)
html = page.content
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find('table')

print(table)
```

```python
for i in table.find_all('td', {"align":'center'}):
    print(i)
```

```python
df = pd.read_csv('annual_inflation_rates_by_months_2024-2000.csv')

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '',text)
```

```python
for i in table.find_all('tr', {'height': '17'}):
    y = i.find_all('td', {'align':'right'})
    
    if(y != [] or y != 'Year'):
        print(y)
```
